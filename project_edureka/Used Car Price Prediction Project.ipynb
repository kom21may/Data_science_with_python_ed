```python
plt.hist(car['year'], bins=20)
plt.xlabel('Manufacturing Year')
plt.ylabel('Frequency')
plt.title('Distribution of Manufacturing Years')
plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=car,x=car['brand'])
plt.xlabel('Manufacturing brand')
plt.ylabel('Count o per brand')
plt.title('Distribution of Brand')
plt.xticks(rotation=45)
plt.show()
```

```python
car['full_model_name'].value_counts()
```

```python
#'full_model_name' 
plt.figure(figsize=(12, 6))
sns.countplot(data=car,x=car['full_model_name'],order=car['full_model_name'].value_counts().index)
plt.xlabel('Manufacturing Model Name')
plt.ylabel('Count o per Model')
plt.title('Distribution of Model')
plt.xticks(rotation=45)
plt.show()
```

```python

# 'model_name'
plt.figure(figsize=(12, 6))
sns.countplot(data=car, x='model_name', order=car['model_name'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Model Name')
plt.ylabel('Count')
plt.title('Distribution of Model Names')
plt.show()
```
