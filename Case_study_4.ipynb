```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('BigMartSalesData.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data['Month'].unique()
```

```python
#1.Plot Total Sales Per Month for the Year 2011. How the total sales have increased over months in the Year 2011. Which month has the lowest Sales?
```

```python
xd=data[data['Year']==2011]
x_axis=xd.groupby('Month')['Quantity'].sum()
```

```python
x_axis
```

```python
new_x=x_axis.reset_index(name='Total_sales')
```

```python
new_x.head()
```

```python
plt.plot(new_x['Month'],new_x['Total_sales'],marker='o')
```

```python
#2.Plot Total Sales Per Month for the Year 2011 as Bar Chart.  Is Bar Chart Better to visualize than a Simple Plot?
plt.bar(new_x['Month'],new_x['Total_sales'])
# here 
```

A bar chart is particularly useful for showing comparisons between discrete categories, like months in this case. It provides a clear visual representation of the sales distribution for each month.

On the other hand, a simple plot (line plot) would typically be used to show trends or changes over continuous or sequential data points. If you're interested in visualizing how the total sales change over time (months), a line plot could be more appropriate.

3.Plot Pie Chart for the Year 2011 Country Wise. Which Country contributes the highest towards sales?


```python
yr_2011=data[data['Year']==2011]
pie_country=yr_2011.groupby('Country')['Quantity'].sum()
```

```python
pie_country
```

```python
pie_axis=pie_country.reset_index(name='Total_sales_country_wise')
```

```python

plt.pie(pie_axis['Total_sales_country_wise'],labels=pie_axis['Country'],autopct="%1.1f%%")
```

4.Plot Scatter Plot for the invoice amounts and see the concentration of the amount.  In which range most of the invoice amounts are concentrated

```python
plt.scatter(range(len(data)),data['Amount'],color='blue', alpha=0.7)
```

```python

# Calculate the concentration range
mean_amount = np.mean(data['Amount'])
std_amount = np.std(data['Amount'])

lower_limit = mean_amount - 2 * std_amount
upper_limit = mean_amount + 2 * std_amount

print(f"Most of the invoice amounts are concentrated in the range: {lower_limit} to {upper_limit}")
```

```python
#3.In scatter plot change the color of Scatter Points
plt.scatter(range(len(data)),data['Amount'],color='red', alpha=0.7)
```

```python
#In Pie Chart Play With Parameters shadow=True, startangle=90,and see how different the chart looks

plt.pie(pie_axis['Total_sales_country_wise'],labels=pie_axis['Country'],autopct="%1.1f%%",startangle=30,shadow=True)
```
