You are given a dataset, which is present in the LMS, containing the number of hurricanes occurring in the United States along the coast of the Atlantic. Load the data from the dataset into your program and plot a Bar Graph of the data, taking the Year as the x-axis and the number of hurricanes occurring as the Y-axis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
hur=pd.read_csv('Hurricanes.csv')
```

```python
hur.head()
```

```python
plt.bar(hur['Year'],hur['Hurricanes'],color='maroon')
plt.xlabel('Year')
plt.ylabel('No of Hurricanes')
plt.title('Year vs Hurricanes')
```

2. .The dataset given, records data of city temperatures over the years2014 and 2015. Plot the histogram of the temperatures over this period for the cities of San Francisco and Moscow

```python
city_temp=pd.read_csv('CityTemps.csv')
city_temp.head()
```

```python
moscow_data=city_temp[(city_temp['Year']==2014) | (city_temp['Year']==2015)]['Moscow']
```

```python
San_data=city_temp[(city_temp['Year']==2014) | (city_temp['Year']==2015)]['San Francisco']
```

```python
plt.hist(moscow_data,bins=20,alpha=0.7,label="Moscow")
plt.hist(San_data,bins=20,alpha=0.7,label="San Francisco")
plt.xlabel('Tempearture')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
