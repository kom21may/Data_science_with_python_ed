2.Data Manipulation:

a.Create a new column -’total_benefitted’ 
that is a sum of inmates benefitted through all modes.[Hint: Use sum method with appropriate axis]

b.Create a new row -“totals” that is the sum of all inmates benefitted through each mode across all states

```python
d6['Total_benefitted']=d6['No. of Inmates benefitted by Adult Education']+d6['No. of Inmates benefitted by Elementary Education']+d6['No. of Inmates benefitted by Higher Education']+d6['No. of Inmates benefitted by Computer Course']
```

```python
d6.head()
```

```python
#b.Create a new row -“totals” that is the sum of all inmates benefitted through each mode across all states
new_row = {
    'STATE/UT': 'All States',
    'YEAR': 'All Years'
}

# Calculate the sum of each column (excluding 'STATE/UT' and 'YEAR') and add to the new row
for column in d6.columns[2:]:
    new_row[column] = d6[column].sum()

new_row_df = pd.DataFrame([new_row])

# Concatenate the original DataFrame with the new row DataFrame
d6 = pd.concat([d6, new_row_df], ignore_index=True)
d6.tail()
```

```python
import matplotlib.pyplot as plt
d6_sorted = d6.sort_values(by="Total_benefitted", ascending=False)
plt.figure(figsize=(10,6))
plt.bar(d6_sorted['STATE/UT'],d6_sorted['Total_benefitted'])
plt.xlabel("State")
plt.ylabel("Total Benefited Inmates")
plt.title("Total Benefited Inmates by State")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
State_max_beneficiary=d6_sorted.iloc[1]['STATE/UT']
```
